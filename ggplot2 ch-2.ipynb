```r
ggplot(mpg, aes(displ, hwy)) +
geom_point() +
facet_wrap(~cyl)
```

```r
ggplot(mpg, aes(displ, hwy)) +
geom_point() + 
geom_smooth(method='lm', formula=y~poly(x,2))
# methods that can be used in geom_smooth:
# loess, rlm, lm, gam.
# ploynomials also can be fitted in the formula
```

```r
# avoid overplotting of points with same values with geom_jitter()
# jitter adds a little random noise to avoid overplotting of points
ggplot(mpg, aes(drv,hwy, color=drv, shape=drv)) +
geom_jitter()
```

```r
ggplot(mpg, aes(drv,hwy, colour=drv)) +
geom_boxplot()
```

```r
ggplot(mpg, aes(drv,hwy, color=drv)) +
geom_violin()
```

```r
ggplot(mpg, aes(hwy)) +
geom_histogram(binwidth=2.5, color='blue') +
geom_freqpoly(, color='red')
```

```r
ggplot(mpg, aes(hwy)) +
geom_freqpoly(binwidth=1, color='green4')
```

```r
ggplot(mpg, aes(hwy, color=drv)) +
geom_freqpoly(binwidth=1)
```

```r
ggplot(mpg, aes(hwy, fill=drv)) + 
geom_histogram(binwidth=0.5) +
facet_wrap(~drv, ncol=1)
```

```r
ggplot(mpg, aes(manufacturer)) +
geom_bar()
```

```r
drugs = data.frame(drug=c('a','b','c'),
                effect=c(4.2, 9.7, 6.1))
```

```r
ggplot(drugs, aes(drug, effect)) +
geom_bar(stat='identity')
```

```r
ggplot(drugs, aes(drug, effect)) + 
geom_point()
```
